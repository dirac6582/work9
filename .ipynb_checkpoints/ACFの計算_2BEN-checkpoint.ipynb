```python
from  ase.io.espresso import  read_espresso_out
import numpy as np
import inspect
```

```python
# 構造データの読み込み
diff_config=np.load("/Users/amanotomohito/works/research/work9/2ben_STR/diff_structure.npy")
```

```python
# BECデータの読み込み．
num_data=11
NUM_ATOM=24


# BECデータの読み込み．
BECarray=np.zeros((num_data,NUM_ATOM,3,3))

for i in range(0,num_data):
    # zero padding
    count_zero = str(i).zfill(2)
    print("STRUCTURE:: ", count_zero)
    BECarray[i-1,:,:,:]=np.load("2ben_BEC/"+count_zero+".npy")
    #print(BECarray[i,:,:,:])
```

```python
print(diff_config)
```

```python
# BECarrayとdiff_configの形を合わせる必要あり．

# for debug
print("diff_config_shape:: ", diff_config.shape)

#diff_config2=diff_config[0:401,:,:]
#print(diff_config2.shape)

# for debug 
#000成分（x軸,y軸,atom ID）を列挙
print(BECarray[0:100,0,0,0])
print("BECarray_shape:: ", BECarray.shape)
```

```python
# 矩形積分によってBEC*座標を足し上げて

dipole2=[]


for timestep in range (0,num_data):
    #timestepでのBECを取り出す.
    BEC_timestep=BECarray[timestep,:,:,:]
    #print(BEC_timestep)
    print("")
    #timestepでのdiff_configを取り出す．
    config_timestep=diff_config[timestep,:,:]
    #print(config_timestep)
    sum=np.zeros(3)
    print(sum)
    for kappa in range (0,NUM_ATOM):
        sum+=np.dot(BEC_timestep[kappa,:,:],config_timestep[kappa,:])
    print(sum)
    dipole2.append(sum)
    
    #print(BEC_timestep[0,:,:])
    ##print(config_timestep[0,:])
    # print(np.dot(BEC_timestep[0,:,:],config_timestep[0,:].T))
    #print(np.dot(BEC_timestep[0,:,:],config_timestep[0,:]))
    
    #print(np.dot(BEC_timestep,config_timestep))
    




# ベンゼンの誘電率. 2.28
```

```python
# ACF の図示
# tupple to numpy
dipole2=np.array(dipole2)

print(dipole2)

```

```python
## 静的誘電率

# x成分
mean_x=np.mean(dipole2[:,0])
var_x=np.mean(dipole2[:,0]**2)

# y成分
mean_y=np.mean(dipole2[:,1])
var_y=np.mean(dipole2[:,1]**2)

# z成分
mean_z=np.mean(dipole2[:,2])
var_z=np.mean(dipole2[:,2]**2)

print(mean_x)
print(mean_y)
print(mean_z)
print(var_x)
print(var_y)
print(var_z)
```

```python
# 誘電率の計算
epsilon_0=8.85418*1e-12
k_b=1.380649*1e-23
e=1.602*1e-19
ang=1e-10
a_0=10.5889
T=300
epsilon_x=(var_x-mean_x**2)*e**2*ang**2/(3*epsilon_0*a_0**3*ang**3*k_b*T)
epsilon_y=(var_y-mean_y**2)*e**2*ang**2/(3*epsilon_0*a_0**3*ang**3*k_b*T)
epsilon_z=(var_z-mean_z**2)*e**2*ang**2/(3*epsilon_0*a_0**3*ang**3*k_b*T)
print(epsilon_x)
print(epsilon_y)
print(epsilon_z)
```

## 積分を飛ばすことができるだろうか？

```python
dipole=[]

# kizami
kizami=4

for timestep in range(0,4000,kizami):
    print(timestep)
    
    ## BECはtimestepで取り出せば良い．
    BEC_timestep=BECarray[timestep,:,:,:]
    #print(BEC_timestep)
    print("")
    
    # 座標
    new_config=np.sum(diff_config[timestep:timestep+kizami-1,:,:],axis=0)
    #print(new_config)
    #print(config_timestep)
    sum=np.zeros(3)
    print(sum)
    for kappa in range (0,96):
        sum+=np.dot(BEC_timestep[kappa,:,:],new_config[kappa,:])
    print(sum)
    dipole.append(sum)
    
    #print(BEC_timestep[0,:,:])
    ##print(config_timestep[0,:])
    # print(np.dot(BEC_timestep[0,:,:],config_timestep[0,:].T))
    #print(np.dot(BEC_timestep[0,:,:],config_timestep[0,:]))
    
    #print(np.dot(BEC_timestep,config_timestep))
    


```
